## Task 2: The Three Branches of Machine Learning

### 1️⃣ Supervised Learning
**Definition:**  
Supervised Learning is used when the dataset contains labels (correct answers). The model learns from input-output pairs and tries to predict the correct output for new, unseen data.

**Examples:**  
- Spam Email Detection  
- House Price Prediction  

---

### 2️⃣ Unsupervised Learning
**Definition:**  
Unsupervised Learning is used when the dataset has no labels. The model explores the data and finds hidden patterns, relationships, or clusters on its own.

**Examples:**  
- Customer Segmentation  
- Grouping Similar News Articles  

---

### 3️⃣ Reinforcement Learning
**Definition:**  
Reinforcement Learning is where an agent learns by interacting with an environment and receiving rewards or penalties. The goal is to maximize rewards over time.

**Examples:**  
- AI Playing Chess  
- Robot Learning to Walk  

---

## Comparison Table

| Real-World Problem            | ML Type               | Why |
|-------------------------------|-----------------------|------|
| Spam Detection                | Supervised Learning   | Has labeled data (spam/not spam) |
| House Price Prediction        | Supervised Learning   | Has actual price values as labels |
| Customer Segmentation         | Unsupervised Learning | No labels; finding hidden groups |
| News Article Clustering       | Unsupervised Learning | Detecting natural groupings |
| AI Playing Chess              | Reinforcement Learning| Learns from rewards/penalties |
| Robot Navigation              | Reinforcement Learning| Improves actions via feedback |

---

### Why This Matters
Choosing the wrong learning type is a common beginner mistake. If you pick the wrong framework, your model will struggle or completely fail before you even start coding.
