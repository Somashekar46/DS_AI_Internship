## Machine Learning (ML): 
Machine Learning is a way to teach computers to learn from data and improve their performance over time, instead of explicitly programming rules. The computer identifies patterns, makes predictions, or takes actions based on what it has learned.

## Types of Machine Learning

1. Supervised Learning:

Definition: The computer learns from data that already has the correct answers (labels). It uses these examples to make predictions on new, unseen data.

Examples: Predicting house prices based on features like size and location, detecting spam emails, or identifying credit card fraud.

2. Unsupervised Learning:

Definition: The computer looks for hidden patterns, structures, or groupings in data that doesn’t have labeled answers. It organizes and summarizes the information by itself.

Examples: Segmenting customers based on buying behavior, clustering similar news articles, or detecting unusual transactions.

3. Reinforcement Learning:

Definition: The computer learns by interacting with an environment and receiving feedback (rewards or penalties). It improves its strategy over time to maximize positive outcomes.

Examples: Teaching a robot to walk, training an AI to play chess or video games, and developing self-driving car navigation systems.

### Traditional Programming vs. Machine Learning

**Traditional Programming:**  
- Humans write explicit rules and logic.  
- Input: Data + Rules = Output: Result.  
- Example: Spam filter using rules like “if subject contains ‘win’, mark as spam.”  

**Machine Learning:**  
- Computers learn patterns from data instead of fixed rules.  
- Input: Data + Expected Output = Output: Prediction or decision.  
- Example: Spam filter trained on thousands of labeled emails to detect spam automatically.  

**Key Difference:**  
- Traditional: Humans tell the computer exactly what to do.  
- ML: Computers figure out patterns and rules by learning from data.
